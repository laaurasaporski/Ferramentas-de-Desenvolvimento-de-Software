```python
# Função para analisar um jogo específico do PS3 em outras plataformas
def comparar_jogo_ps3(nome_jogo):
    jogo_todas_plat = periodo_modelo[periodo_modelo['name'] == nome_jogo]
    
    if len(jogo_todas_plat) > 1:
        print(f"\n=== {nome_jogo} ===")
        
        # Ordenar por vendas para ver o ranking
        jogo_ordenado = jogo_todas_plat.sort_values('total_sales', ascending=False)
        
        for _, row in jogo_ordenado.iterrows():
            destaque = "🎯 PS3" if row['platform'] == 'PS3' else "   "
            print(f"{destaque} {row['platform']}: {row['total_sales']:.2f} milhões")
        
        # Posição do PS3 no ranking
        ps3_posicao = jogo_ordenado.reset_index().index[jogo_ordenado['platform'] == 'PS3'].tolist()
        if ps3_posicao:
            print(f"PS3 ficou em {ps3_posicao[0] + 1}º lugar de {len(jogo_ordenado)} plataformas")
        
        return jogo_ordenado
    return None

# Vamos analisar os top 5 jogos do PS3 em outras plataformas
top_jogos_ps3 = ps3_data.nlargest(5, 'total_sales')['name'].tolist()
print("Analisando os TOP 5 jogos do PS3:")
print(top_jogos_ps3)

for jogo in top_jogos_ps3:
    comparar_jogo_ps3(jogo)
```

```python
# Calcular proporções de jogos "M" por região
print("=== PROPORÇÃO DE JOGOS 'M' POR REGIÃO ===")

# Vendas médias por jogo
vendas_media_e = {
    'NA': vendas_e['NA'] / len(jogos_e),
    'EU': vendas_e['EU'] / len(jogos_e), 
    'JP': vendas_e['JP'] / len(jogos_e)
}

vendas_media_m = {
    'NA': vendas_m['NA'] / len(jogos_m),
    'EU': vendas_m['EU'] / len(jogos_m),
    'JP': vendas_m['JP'] / len(jogos_m)
}

print(f"\nVendas médias por jogo 'E':")
print(f"  NA: {vendas_media_e['NA']:.3f} milhões/jogo")
print(f"  EU: {vendas_media_e['EU']:.3f} milhões/jogo") 
print(f"  JP: {vendas_media_e['JP']:.3f} milhões/jogo")

print(f"\nVendas médias por jogo 'M':")
print(f"  NA: {vendas_media_m['NA']:.3f} milhões/jogo")
print(f"  EU: {vendas_media_m['EU']:.3f} milhões/jogo")
print(f"  JP: {vendas_media_m['JP']:.3f} milhões/jogo")

# Calcular ratio M/E para cada região
ratio_na = vendas_media_m['NA'] / vendas_media_e['NA']
ratio_eu = vendas_media_m['EU'] / vendas_media_e['EU'] 
ratio_jp = vendas_media_m['JP'] / vendas_media_e['JP']

print(f"\nRatio vendas M/E (quanto jogos M vendem comparado aos E):")
print(f"  NA: {ratio_na:.2f}x")
print(f"  EU: {ratio_eu:.2f}x") 
print(f"  JP: {ratio_jp:.2f}x")
```

No Japão, os jogos de classificação "M" vendem menos do que em outros paises

```python
# Primeiro, preparar os dados para os testes de hipóteses
import scipy.stats as stats

# Definir alfa
alpha = 0.05
print(f"Nível de significância escolhido: α = {alpha}")

# Verificar se tem dados suficientes para os testes
print(f"\nDados disponíveis no período 2006-2016: {len(periodo_modelo)} jogos")
print(f"Jogos com user_score válido: {periodo_modelo['user_score'].notna().sum()}")
```

```python
# Filtrar dados para Xbox One e PC
xbox_data = periodo_modelo[periodo_modelo['platform'] == 'XOne']['user_score'].dropna()
pc_data = periodo_modelo[periodo_modelo['platform'] == 'PC']['user_score'].dropna()

print("=== TESTE 1: XBOX ONE vs PC ===")
print(f"Xbox One: {len(xbox_data)} jogos")
print(f"PC: {len(pc_data)} jogos")

if len(xbox_data) > 0 and len(pc_data) > 0:
    print(f"Média Xbox One: {xbox_data.mean():.2f}")
    print(f"Média PC: {pc_data.mean():.2f}")
else:
    print("⚠️ Dados insuficientes para o teste")
```

```python
t_stat, p_value = stats.ttest_ind(xbox_data, pc_data)

print("=== RESULTADOS DO TESTE T ===")
print(f"Estatística t: {t_stat:.4f}")
print(f"P-valor: {p_value:.6f}")
print(f"Alfa: {alpha}")

# Interpretação
print(f"\n=== INTERPRETAÇÃO ===")
if p_value < alpha:
    print(f"✅ REJEITAMOS H₀ (p = {p_value:.6f} < {alpha})")
    print("Há diferença significativa entre Xbox One e PC")
else:
    print(f"❌ NÃO REJEITAMOS H₀ (p = {p_value:.6f} >= {alpha})")
    print("Não há diferença significativa entre Xbox One e PC")


diferenca = pc_data.mean() - xbox_data.mean()
print(f"\nDiferença das médias: {diferenca:.3f} pontos")
print(f"PC tem média {diferenca:.3f} pontos maior que Xbox One")
```

```python
# Filtrar dados para Action e Sports
action_data = periodo_modelo[periodo_modelo['genre'] == 'Action']['user_score'].dropna()
sports_data = periodo_modelo[periodo_modelo['genre'] == 'Sports']['user_score'].dropna()

print("=== TESTE 2: ACTION vs SPORTS ===")
print(f"Jogos de Action: {len(action_data)} jogos")
print(f"Jogos de Sports: {len(sports_data)} jogos")

if len(action_data) > 0 and len(sports_data) > 0:
    print(f"Média Action: {action_data.mean():.2f}")
    print(f"Média Sports: {sports_data.mean():.2f}")
    print(f"Desvio padrão Action: {action_data.std():.2f}")
    print(f"Desvio padrão Sports: {sports_data.std():.2f}")
else:
    print("⚠️ Dados insuficientes para o teste")
```

```python
t_stat_2, p_value_2 = stats.ttest_ind(action_data, sports_data)

print("=== RESULTADOS DO TESTE 2: ACTION vs SPORTS ===")
print(f"Estatística t: {t_stat_2:.4f}")
print(f"P-valor: {p_value_2:.6f}")
print(f"Alfa: {alpha}")

# Interpretação
print(f"\n=== INTERPRETAÇÃO ===")
if p_value_2 < alpha:
    print(f"✅ REJEITAMOS H₀ (p = {p_value_2:.6f} < {alpha})")
    print("Há diferença significativa entre Action e Sports")
else:
    print(f"❌ NÃO REJEITAMOS H₀ (p = {p_value_2:.6f} >= {alpha})")
    print("Não há diferença significativa entre Action e Sports")

diferenca_2 = action_data.mean() - sports_data.mean()
print(f"\nDiferença das médias: {diferenca_2:.3f} pontos")
print(f"Action tem média {diferenca_2:.3f} pontos {'maior' if diferenca_2 > 0 else 'menor'} que Sports")
```

1. Teste bilateral: Usei testes de duas caudas porque não tinha uma expectativ prévia sobre qual grupo seria superior

2. Comparação de médias: Como queria comparar avaliações médias entre dois grupos independentes, o teste t de Student foi apropriado

3. Hipótese conservadora: A hipótese nula sempre assume "não há diferença" - isso força a evidência estatística a ser forte para rejeitá-la

Por que escolhi α = 0.05:É o nível mais comum em pesquisas
